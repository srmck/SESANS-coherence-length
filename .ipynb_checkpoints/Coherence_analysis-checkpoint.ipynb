```python
import numpy as np
from scipy.signal import hilbert,convolve,savgol_filter,sawtooth,find_peaks,correlate,square
from scipy.fft import fft, fftfreq, ifft
from scipy.integrate import simpson
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import csv 
#from astropy.convolution import convolve

np.set_printoptions(suppress=True)

SMALL_SIZE = 20
MEDIUM_SIZE = 25
BIGGER_SIZE = 30

plt.rcParams['figure.figsize'] = (16, 10)
plt.rc('lines', linewidth=3)

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

def splitter(x,lim):
    """Splits data into peaks. x is the list of points above some thresold."""
    N = 1  #number of splits
    diffs = np.diff(x)
    for i in diffs:
        if i>lim: N += 1
    temp = [[] for n in range(N)]
    
    tempN = 0
    for i in range(len(x)-1):
        if diffs[i]<=lim:
            temp[tempN].append(x[i])
        else:
            temp[tempN].append(x[i])
            tempN += 1
    filtered = temp.copy()
    for i in temp:
        if len(i) <= 3: filtered.remove(i)
    return filtered

def gauss(x,a,s):
    return 1/(np.sqrt(2*np.pi)*s)*np.exp(-(x-a)**2/2/s**2)

def cos(x,a,f,s):
    return s+a*np.cos(x*f)

def ecos(x,a,f,p,s):
    return s+a*np.cos(np.sqrt(x)*f+p)

def ecos_nophase(x,a,f,s):
    return s+a*np.cos(np.sqrt(x)*f)

def line(x,a,b):
    return a*x+b

def quad(x,a,b):
    return 1-a*x**2-b*x

def exp_sum(x,amps,avs,sigs):
    return np.sum([amps[i]*np.exp(-(x-avs[i])**2/2/sigs[i]**2) for i in range(len(avs))],axis=0)

def ampgauss(x,amp,a,s):
    return amp*np.exp(-(x-a)**2/2/s**2)

def mod_quad(x,a,b,c):
    return a*x**2 + b*x + c

def asyg(t,amp,av,sig,a):
    scale = sig/(1+np.exp(a*(t-av)))
    return amp*np.exp(-(t-av)**2/(2*scale**2))

G = -1.832471e8    #gyromagnetic ratio of neutron (rad/T/m^2)
M = 1.674927e-27    #mass of neutron (kg)
HBAR = 1.054571e-34    #plancks constant (kg m^2/s)
H = HBAR*2*np.pi
CL = G/(HBAR*2*np.pi)*M    #Larmor phase constant (T^-1 m^-1 AA^-1)
print(f'Larmor phase constant: {round(CL)}')

#Background function constant
temp_freq = 3
spc = 2*1.67e-27*temp_freq*1e6*1.2/6.626e-34/np.tan(40/180*np.pi)*1e-10 #per Angstrom
print(f'a = {round(spc)} A^-1 at rf freq {temp_freq} MHz')
sld = 2.06e-6  #per Anstrom squared
d = 10.5*1e4
tc = sld*d/np.sqrt(spc)
print(f'Cosine freq {round(tc*1e2,5)} um^(-1/2)')
```

## Wavepacket spread

```python
def wp_spread(t,a):
    """Wavepacket spread from gaussian."""
    return a*np.sqrt(1 + (HBAR*t/M/a**2/2)**2)

def get_v(l):
    """Converts wavelength in Angstroms to velocity in m/s."""
    return H/M/l*1e10

L = 10
a0 = 1e-6  #initial wavepacket standard deviation
beam = .01  #beam diameter
time = np.linspace(0,L/get_v(4),200)
print(f'Flight time = {round(time[-1]*1e3,3)} (ms)')

fig,ax = plt.subplots()
ax.ticklabel_format(useOffset=False)
plt.plot(time,2.35*wp_spread(time,a0)*1e6,label='wavepacket size')
#plt.plot(time,np.full(len(time),beam)*1e6,label='beam size')
plt.ylabel('Wavepacket size (um)')
plt.xlabel('Time (s)')
plt.legend()
plt.grid()
plt.show()
```

# Convolution testing

```python
def trapzoid_signal(t, width=2., slope=1., amp=1., offs=0):
    a = slope*width*sawtooth(2*np.pi*t/width, width=0.5)/4.
    a[a>amp/2.] = amp/2.
    a[a<-amp/2.] = -amp/2.
    return a + amp/2. + offs

t = trapzoid_signal(np.linspace(0,2,400),width=2,slope=20)
t = t/simpson(t)
ac = convolve(t,t)
ac = ac/simpson(ac)

plt.plot(t)
plt.plot(ac)
plt.grid()
plt.show()
```

```python
x_g = np.linspace(-1,1,199)
x = np.linspace(0,6,1000)

def cpeak(x,amp,av,sig,width,height):
    s = av-width/2
    e = av+width/2
    p = x[-1]
    

def tri(x,s,e,p,a,norm=False):
    s = np.argmin(np.abs(x-s))
    e = np.argmin(np.abs(x-e))
    p = np.argmin(np.abs(x-p))
    xp = np.linspace(x[0],10*x[-1],10*len(x))
    m = a/(xp[e]-xp[s])
    c = (e-s)//2
    temp = np.zeros(len(x))
    for indx,val in enumerate(x):
        n = indx//p
        if indx%p>s and indx%p<(c+s+1):
            temp[indx] = m*(val-xp[s+n*p])
        elif indx%p>(c+s) and indx%p<e:
            temp[indx] = -m*(val-xp[e+n*p])
    if norm == True: temp = temp/simpson(temp,x)
    return temp

s,e = 3-1.44/2,3+1.44/2
p = 10*(s+e)
t = tri(x,s,e,p,1,norm=True)
g1 = gauss(x_g,0,.1)
cvd = convolve(t,g1,boundary=None)

fig,ax = plt.subplots()
plt.plot(x,t)
plt.plot(x,cvd)
#plt.plot(x_g,g1)
plt.grid()
plt.show()
```
